```python
questions = ["1. How many people have heart disease and how many doesn't",
             "2. People of which gender most have heart disease.",
             "3. People of which gender has which type of chest pain",
             " 4.People with which chest pain are prone to heart disease. "]
```

```python
#first question

df.target.value_counts()
```

```python
df.target.value_counts().plot(kind ='bar', color=['red','black'])
plt.title("Heart disease values")
plt.xlabel('1 = Heart disease,0=no heart disease')
plt.ylabel('amount');
```

```python
df.target.value_counts().plot(kind = 'pie', figsize=(8,6))
plt.legend(['disease','no disease'])
```

```python
#second question
df.sex.value_counts()
```

```python
df.sex.value_counts().plot(kind ='pie')
plt.title("male female ratio")
plt.legend(['male','female'])
```

```python
pd.crosstab(df.target,df.sex)
```

```python
#question 3
df.cp.value_counts()
```

```python
df.cp.value_counts().plot(kind='bar', color=['red','black','blue','yellow'])
plt.title('Chest pain vs count')
```

```python
pd.crosstab(df.sex,df.cp).plot(kind='bar',color=['red','black','blue','yellow'])
plt.xlabel('0 = female, 1=male')
plt.title("type of chest pain for sex")
```

```python
pd.crosstab(df.cp,df.target)
```

```python
pd.crosstab(df.cp,df.target).plot(kind='bar')
```
